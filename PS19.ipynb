# CSV --> Comma Seperated Values

```python
data = """Name,Age,Country
John,25,USA
Alice,30,Canada
Bob,28,UK"""

lines=data.split('\n')
# print(lines)
keys=lines[0].split(',')
# print(keys)
result=[]
for line in lines[1:]:
    values=line.split(',')
    # print(values)
    entry=dict(zip(keys,values))
    # print(entry)
    result.append(entry)
result 
```
